### Question 1 - Simple plot

**WRITE** the sin wave code from the slide into the cell below and create a plot...

  1. Can you change the size?
  2. Can you change the DPI?
  

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('classic')
fig = plt.figure(figsize=(4,4), dpi=200)
t = np.arange(0.0, 1.0, 0.01)
s = np.sin(2*np.pi*t)
ax1 = fig.add_subplot(1,1,1)
ax1.set_ylabel('volts')
ax1.set_title('a sine wave')
ax1.plot(t, s+5.0, color='blue', lw=2)
plt.savefig("foo.png",dpi=400)
````

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(10,5))
x = np.random.normal(100,1,100)
bins = 20
ax1 = fig.add_subplot(2,2,1)
ax1.hist(x, normed=True,color="#9999FF",bins=bins,histtype='stepfilled')
ax2 = fig.add_subplot(2,1,2)
ax2.hist(x, normed=True,color="#9999FF",bins=bins,histtype='stepfilled')
ax3 = fig.add_subplot(2,2,2)
ax3.hist(x, normed=True,color="#9999FF",bins=bins,histtype='stepfilled')
plt.subplots_adjust(wspace=0.2,hspace=.4)
ax1.set_ylabel('y-axis')
ax1.set_title('title')
plt.show()
````
